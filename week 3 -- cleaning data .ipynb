# SALARY

```python
df.head()
```

```python
df["salary"]=df["salary"].apply(lambda x:x.replace("k",""))
df.head()
```

# Clean established col

```python
df["established"].replace(-1,np.nan,inplace=True)
df["established"].isnull().sum()
```

```python
df.loc[df["established"].isnull()]
```

### replace these nan values to unknown str

```python
df["established"].fillna("Unknown",inplace=True)
```

```python
df.loc[df["established"]=="Unknown"]
```

# LOCATION

```python
df['location']=df['location'].apply(lambda x:x.replace(","," "))
df.head()
```

```python
df[['new_location','split_location']]=df['location'].str.rsplit(" ",1,expand=True)
```

```python
df.head()
```

```python
df.drop(columns=["split_location","location"],inplace =True)
```

```python
df.rename(columns={"new_location":"location"})
```

```python
df.head()
```

# RATING

```python
df["rating"]=df["rating"].replace(-1.0,np.nan)
```

```python
mean=np.mean(df["rating"])
```

```python
df["rating"]=df["rating"].fillna(mean)
```

```python
df["rating"]=df["rating"].round(1)
```

```python
df.head()
```

# EASY_apply

```python
df["easy_apply"]=df["easy_apply"].replace("-1","FALSE")
```

# Clean chipotle dataset

```python
data=pd.read_csv("chipotle.tsv",sep="\t")
```

```python
data.head()
```

```python
data.tail()
```

```python
data.shape
```

```python
data.info()
```

## change data type of item_price

```python
data["item_price"]=data["item_price"].apply(lambda x:x.replace("$",""))
```

```python
data["item_price"]=data["item_price"].astype(float).round(1)
```

```python
data.head()
```

```python
data.isnull().sum()
```

## 1. Quantity of each item

```python
data.head()
```

```python
data["item_name"].value_counts()
```

## 2.	What was the most ordered item in the choice_description column? 

```python
data.head()
```

```python
data.groupby("choice_description")["quantity"].sum().idxmax()
```

## 3.	How many items were ordered in total?

```python
Total_item_ordered=np.sum(data["quantity"])
Total_item_ordered
```

## 4.	How much was the revenue for the period in the dataset?

```python
data.head()
```

```python
revenue_generated=np.sum(data["item_price"])
revenue_generated.round(1)
```

## 5.	How many orders were made in the period?

```python
Total_orders=data["order_id"].nunique()
Total_orders
```

## 6.	What is the average revenue amount per order?

```python
data.head(10)
```

```python
data["revenue"]=data["quantity"]*data["item_price"]
data.head()
```

```python
sum_revenue_per_order=data.groupby("order_id")["revenue"].sum()
sum_revenue_per_order.mean()
```
